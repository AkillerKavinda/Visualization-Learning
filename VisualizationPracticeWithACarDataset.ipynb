# Data cleaning

```python
df.isnull().sum()
```

```python
df = df.fillna(df.mean(numeric_only=True))
```

```python
df.isnull().sum()
```

```python
for col in ['Make', 'Model', 'Type', 'Origin', 'DriveTrain']:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df.isnull().sum()
```

```python
df[['MSRP', 'Invoice']] = df[['MSRP', 'Invoice']].fillna(df[['MSRP', 'Invoice']].mode())
```

```python
df.isnull().sum()
```

# Remove all the records where weight is greater than 4000

```python
df.head(1)
```

```python
df = df[df['Weight']<4000]
```

# Increase all the values of 'MPG_City' column by 3

```python
df['MPG_City'] = df['MPG_City'].apply(lambda x:x+3)
```
